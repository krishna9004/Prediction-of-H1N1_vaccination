## LOGISTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
X=df.drop("h1n1_vaccine",axis=1)
y=df["h1n1_vaccine"]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=1)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
predictions=model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
```

```python
from sklearn import metrics
cm=metrics.confusion_matrix(y_test,predictions,labels=[1,0])
df_cm=pd.DataFrame(cm,index=[i for i in["1","0"]],
                  columns=[i for i in ["Predict 1","Predict 0"]])
plt.figure(figsize=(7,5))
sns.heatmap(df_cm,annot=True,fmt='g')
```

```python
print(metrics.classification_report(y_test,predictions))
```

## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
model2.fit(X_train,y_train)
print(model2.score(X_train,y_train))
print(model2.score(X_test,y_test))
```

## the model is overfitted 

```python
model2=DecisionTreeClassifier(max_depth=3,criterion='entropy')
model2.fit(X_train,y_train)
print(model2.score(X_train,y_train))
print(model2.score(X_test,y_test))
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from scipy.stats import zscore 
```

```python
model3=KNeighborsClassifier()
model3.fit(X_train,y_train)
predicted_labels=model3.predict(X_test)
print(model3.score(X_test,y_test))
print(model3.score(X_train,y_train))
```

## Naive Model

```python
diab_model.fit(X_train,y_train)
```

## Bagging

```python
from sklearn.ensemble import BaggingClassifier
model_bag=BaggingClassifier(n_estimators=75,base_estimator=model2)
model_bag.fit(X_train,y_train)
print(model_bag.score(X_train,y_train))
print(model_bag.score(X_test,y_test))
```

## Boosting

```python
from sklearn.ensemble import AdaBoostClassifier
model_bag=AdaBoostClassifier(n_estimators=1)
model_bag.fit(X_train,y_train)
print(model_bag.score(X_train,y_train))
print(model_bag.score(X_test,y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
model_bost=GradientBoostingClassifier(n_estimators=1)
model_bost.fit(X_train,y_train)
print(model_bost.score(X_train,y_train))
print(model_bost.score(X_test,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
model_forest=RandomForestClassifier(max_depth=3)
model_forest.fit(X_train,y_train)
print(model_forest.score(X_train,y_train))
print(model_forest.score(X_test,y_test))
```
